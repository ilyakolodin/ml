добавим новый признак. Сколько секунд осталось в целом. 

```python
data['seconds_remaining_total'] = 60*data['minutes_remaining'] + data['seconds_remaining']
```

```python
data.columns
```

Выделим количественные признаки 

```python
kol = data[['loc_x', 'loc_y', 'shot_distance', 'seconds_remaining_total', 'period']]
kol.head()
```

Проводим нормализацию 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
kol = pd.DataFrame(sc.fit_transform(kol))
kol.head()
```

```python
y = data[target]
categ = data.drop(['loc_x', 'loc_y', 'shot_distance', 'seconds_remaining_total','period', target], axis = 1)
```

```python
categ = pd.get_dummies(categ)
categ.head()
```

```python
categ = categ.reset_index(drop = True)
kol = kol.reset_index(drop = True)
```

```python
X = pd.concat([kol, categ], axis=1)
```

```python
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
def get_score(X,y, random_seed=42, model='lr'):
    if model is 'lr':
        model = LogisticRegression()
    if model is 'Lasso':
        model = LogisticRegression(penalty='l1', solver = 'liblinear')
    if model is 'Ridge':
        model = LogisticRegression(penalty='l2')
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_seed )  
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
def get_coef(X,y, random_seed=42, model='lr'):
    if model is 'lr':
        model = LogisticRegression()
    if model is 'Lasso':
        model = LogisticRegression(penalty='l1', solver = 'liblinear')
    if model is 'Ridge':
        model = LogisticRegression(penalty='l2')
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_seed )  
    model.fit(X_train, y_train)
    return model.coef_
```

```python
lr_score = get_score(X,y)
ridge_score = get_score(X,y, model='Ridge')
lasso_score = get_score(X,y, model='Lasso')
```

```python
lr_score
```

```python
lasso_score
```

```python
ridge_score
```

```python
lr_coef = get_coef(X,y)
lasso_coef = get_coef(X,y, model='Lasso')
ridge_coef = get_coef(X,y, model='Ridge')
```

```python
coef_data = pd.concat([pd.Series(lr_coef.ravel()), pd.Series(lasso_coef.ravel()), pd.Series(ridge_coef.ravel())], axis=1)
coef_data['index'] = coef_data.index
coef_data = coef_data[coef_data['index']<6]
coef_data.rename(columns = {0:'LinearRegression', 1:'Lasso', 2:'Ridge'}, inplace = True)
```

```python
coef_data.plot(x='index', y=[0, 1, 2], kind="bar")
```

На граффике изображены в сравнении первые 5 коэффициентов для разных моделей. Простая модель линейной регрессии дает в среднем большие коэффициенты.

```python
from sklearn.metrics import roc_auc_score
```

```python
def get_roc(X,y, random_seed=42, model='lr'):
    if model is 'lr':
        model = LogisticRegression()
    if model is 'Lasso':
        model = LogisticRegression(penalty='l1', solver = 'liblinear')
    if model is 'Ridge':
        model = LogisticRegression(penalty='l2')
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_seed )  
    model.fit(X_train, y_train)
    return roc_auc_score(y_test, model.predict_proba(X_test)[:, 1]), roc_auc_score(y_test, model.decision_function(X_test))

print(get_roc(X,y))
print(get_roc(X,y, model='Lasso'))
print(get_roc(X,y, model='Ridge'))
```
